```python
data = load_dataset('dataBase/train.csv')
dataTest = load_dataset('dataBase/test.csv')

data.dropna(inplace=True)
dataTest.dropna(inplace=True)

data = data.drop(['Unnamed: 0'],axis=1)
dataTest = dataTest.drop(['Unnamed: 0'],axis=1)

data['Gender'].replace({'Male': 0, 'Female': 1}, inplace = True)
data['Customer Type'].replace({'Loyal Customer': 0, 'disloyal Customer': 1}, inplace = True)
data['Type of Travel'].replace({'Personal Travel': 0, 'Business travel': 1}, inplace = True)
data['Class'].replace({'Eco Plus': 0, 'Eco':1, 'Business': 2}, inplace = True)
data['satisfaction'].replace({'neutral or dissatisfied': 0, 'satisfied': 1}, inplace = True)

dataTest['Gender'].replace({'Male': 0, 'Female': 1}, inplace = True)
dataTest['Customer Type'].replace({'Loyal Customer': 0, 'disloyal Customer': 1}, inplace = True)
dataTest['Type of Travel'].replace({'Personal Travel': 0, 'Business travel': 1}, inplace = True)
dataTest['Class'].replace({'Eco Plus': 0, 'Eco':1, 'Business': 2}, inplace = True)
dataTest['satisfaction'].replace({'neutral or dissatisfied': 0, 'satisfied': 1}, inplace = True)

data.head()
```

```python
#Models a impleementar:
    #Regressor logic 
    #SVM
    #MLP (Multiple layer perception)
    #Random forest
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
#Implementacio del moodel de Regressio Logistica

def RegresorLogistic(x,y,particio):
    for size in particio:
        #particio train i test
        x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=size , test_size = 1-size)

        #Configurem el regressor logistic
        logireg = LogisticRegression(C=2.0, fit_intercept=True, penalty='l2')

        #Entrenem el regresssor logistic
        logireg.fit(x_train, y_train)

        #Resultats del accuracy en el test
        print("Accuracy Logistic: ",logireg.score(x_test, y_test), " amb el ", size, " de train")

        
```

```python
particio = [0.5,0.7,0.8,0.9]
x = data1.drop(["satisfaction"],axis=1).values
y = data1["satisfaction"].values

print("Logistic amb data1 i particio sobre el dataset")
RegresorLogistic(x,y,particio)

x = data2.drop(["satisfaction"],axis=1).values
y = data2["satisfaction"].values

print("")
print("Logistic amb data2 i particio sobre el data set")
RegresorLogistic(x,y,particio)
```

```python
from sklearn.preprocessing import StandardScaler

x_train = data2.drop(["satisfaction"],axis=1).values
y_train = data2["satisfaction"].values
x_test = dataTest2.drop(["satisfaction"],axis=1).values
y_test = dataTest2["satisfaction"].values

ss = StandardScaler(with_mean=False,with_std=False).fit(x_train)

#Escalem els atributs
x_train_scaled = ss.transform(x_train)
x_test_scaled = ss.transform(x_test)
```

```python
#Regressor Logistic
logireg = LogisticRegression()

#Entrenem el regresssor logistic
logireg.fit(x_train, y_train)

#Resultats del accuracy en el test
print("Accuracy Logistic: ",logireg.score(x_test, y_test))


#Entrenem el regresssor logistic
logireg.fit(x_train_scaled, y_train)

#Resultats del accuracy en el test
print("Accuracy Logistic dades escalades: ",logireg.score(x_test_scaled, y_test))
```

```python
#Implementacio del model MLP
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier()

mlp.fit(x_train,y_train)

print("Accuracy MLP: ",mlp.score(x_test, y_test))

mlp.fit(x_train_scaled,y_train)

print("Accuracy MLP dades escalades: ",mlp.score(x_test_scaled, y_test))
```

```python
size = 5000
print(x_train[:size-1].shape)
x_mini_train = x_train[:size-1]
y_mini_train = y_train[:size-1]
x_mini_test = x_test[:size-1]
y_mini_test = y_test[:size-1]

x_mini_train_scaled = x_train_scaled[:size-1]
x_mini_test_scaled = x_test_scaled[:size-1]
```

```python
#Implementacio del model de SVM amb les dades reduides
from sklearn.svm import SVC

#Configurem la SVM
svc = SVC(C=10)

#Entrenem el regresssor logistic
svc.fit(x_mini_train, y_mini_train)

#Resultats del accuracy en el test
print("Accuracy SVM",svc.score(x_mini_test, y_mini_test),"amb",size,"files")

#Entrenem el regresssor logistic
svc.fit(x_mini_train_scaled, y_mini_train)

#Resultats del accuracy en el test
print("Accuracy SVM amb dades escalades",svc.score(x_mini_test_scaled, y_mini_test),"amb",size,"files")
```
