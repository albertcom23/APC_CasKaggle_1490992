```python
import numpy as np
import pandas as pd
from sklearn import datasets

def load_dataset(path):
    dataset = pd.read_csv(path, header=0, delimiter=',')
    return dataset
```

```python
train = load_dataset('dataBase/train.csv')
test = load_dataset('dataBase/test.csv')

print("Dimensionalitat del conjunt train:", train.shape)
print("Dimensionalitat del conjunt test:", test.shape)
```

```python
print("Dades nules en el train:", train.isnull().sum())
print("")
print("Dades nules en el test:", test.isnull().sum())
```

Veiem que la unica variable de la nostre base de dades que conté elements sense valor és l'atribut Arrival Delay in Minutes tant en el train com en el test.
